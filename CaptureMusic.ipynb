```python
import numpy as np

import urllib2
import urllib
import requests
from bs4 import BeautifulSoup
import re
```

```python
def getHTML(url):
    hdr = {'User-Agent':'Mozilla/5.0'}
    req = urllib2.Request(url,headers=hdr)
    page = urllib2.urlopen(req).read()

    soup = BeautifulSoup(page, 'html.parser')
    return soup
```
